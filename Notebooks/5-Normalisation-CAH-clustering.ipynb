```python
# Librairie pandas (manipulation de données csv, dataframe, etc.)
import pandas as pd

# Import et lecture du corpus :
corpus = pd.read_csv('corpus_tokmorph.csv')
corpus.head()
```

```python
corpus = corpus.drop('Unnamed: 0', 1)
corpus = corpus.drop('Unnamed: 0.1', 1)
corpus
```

```python
## Calcul de la fréquence de chacun des mots / lemme / pos : 

count = corpus.groupby(['Songwriter(s)', 'words']).size()
count.head
```

```python
pd.DataFrame(count, columns=['freq'])
```
